```python
# Looping all the value and fill up the null value
for col in data.columns:
    if data[col].dtype == 'object':
        print(col)
        data[col].fillna(value=data[col].mode()[0], inplace=True)
    else:
        data[col].fillna(value=data[col].median(), inplace=True)
```

```python
# Check if the null value of the data again with descending order
data.isnull().sum().sort_values()
```

```python
# Rename the column
data.rename(columns={'availability 365':'days_booked'}, inplace=True)
```

```python
# Convert all columns to lowercase and replace the space with underscore
data.columns
```

```python
data.columns =[col.lower().replace(' ','_') for col in data.columns]
data.columns
```

```python
# Check the column name 
data.head(3)
```

```python
# Remove the dollar sign 
data[['price','service_fee']].head()
```

```python
def remove_dollar_comma_sign(value):
    if pd.isna(value):
        return np.Nan
    else:
        return value.replace("$","").replace(",","")
```

```python
data['price'] = data['price'].apply(lambda x:remove_dollar_comma_sign(x))
data['service_fee'] = data['service_fee'].apply(lambda x:remove_dollar_comma_sign(x))
```

```python
data[['price','service_fee']].head()
```

```python
# Data manipulation
# Standardizing capitalization using str.title()
data['neighbourhood_group'] = data['neighbourhood_group'].str.title()

# Manually replacing misspelled words or variations
corrections = {
    'Broklyn': 'Brooklyn',
    'brookln': 'Brooklyn',
    'manhatan': 'Manhattan',
    'manhattan': 'Manhattan',
    'Brookln': 'Brooklyn',
    'Manhatan': 'Manhattan',
}

for incorrect, correct in corrections.items():
    data['neighbourhood_group'] = data['neighbourhood_group'].str.replace(incorrect, correct)
```
